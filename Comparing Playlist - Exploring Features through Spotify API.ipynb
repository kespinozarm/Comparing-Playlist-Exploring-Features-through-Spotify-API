```python
import pandas as pd
```

```python
df = pd.read_csv("spotifyprueba.csv")
```

```python
df.head()
```

```python
df.loc[df['artist_genres'].str.contains("canadian contemporary r&b|canadian pop|pop", case=False, na=False), 'artist_genres'] = 'pop'
df.loc[df['artist_genres'].str.contains("hip hop|rap|slap|alternative", case=False, na=False), 'artist_genres'] = "hip hop"
df.loc[df['artist_genres'].str.contains("rock", case=False, na=False), 'artist_genres'] = "rock"
df.loc[df['artist_genres'].str.contains("reggaeton|urbano latino|boricua|corrido|chill", case=False, na=False), 'artist_genres'] = "urbano latino"
df.loc[df['artist_genres'].str.contains("alt z|indie|melodic|krushclub", case=False, na=False), 'artist_genres'] = "indie"
```

```python
df["artist_genres"].value_counts() 
```

```python
df.dropna(how="all")
```

```python
df_features = pd.read_csv("spotifyfeatures.csv")
```

```python
df_features.head()
```

```python
df_features.rename(columns={"id": "track_id"}, inplace=True)
```

```python
df_combined = pd.merge(df, df_features, left_on='track_id', right_on='track_id', how='inner')
```

```python
df_combined.columns
```

```python
df_features1= df_combined[["artist_name","popularity","artist_genres","danceability","energy","speechiness","instrumentalness","valence"]]
```

```python
df_features1.head()
```

```python
df_features_1 = df_features_1.groupby("artist_genres")[["danceability", "energy", "speechiness", "instrumentalness", "valence"]].agg("mean").reset_index()
```

```python
df_final_1 = pd.concat([df_features1, df_features_1], axis=0)  # Para concatenar filas
```

```python
df_final_1.head()
```

```python
df_final_1["speechiness"].mode().mean()
```
